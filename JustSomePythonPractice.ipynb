```python
import math

# объявление переменных a b c и присвоение им значений
a = 3.4
b = 6.22
c = 0.149

# формула
F = (math.sqrt(a * b))/(b-2*c)

# значение функции
print("Значение функции = ", F)
```

```python
# абсолютная погрешность (нестрогая)
pa = 0.1
pb = 0.01
pc = 0.001

# абсолютная погрешность (строгая!)
pastr = pa/2
pbstr = pb/2
pcstr = pc/2

# частная производная a
pra = math.sqrt(b) / (2 * math.sqrt(a) * (b - 2 * c))
zn = b - 2 * c

# частная производная c
prc = (2*math.sqrt(a*b))/math.pow(zn, 2)

# частная производная b
prb = math.sqrt(a)*(math.pow(zn,-1)*(1/(2*math.sqrt(b)))-math.sqrt(b)* (math.pow(zn, -2)))

# абсолютная погрешность ф-ии
pf = abs(pra) * pa + abs(prb) * pb + abs(prc) * pc # формула
print("Абсолютная погрешность ф-ии = ", pf)
```

```python
# относительная погрешность
fotn = pf / abs(F)
print("Относительная погрешность ф-ии = ", fotn)

# a + дельта a и т.д. (для вычисления экспер. пог – ти)
ad = a + pa
bd = b + pb
cd = c + pc

# функция с ad, bd, cd
FD = (math.sqrt(ad * bd)) / (bd - 2 * cd);
# экспериментальная погрешность
Fexp = abs(F-FD);
print("Значение экспериментальной погрешности = ", Fexp)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def foo(x):
  return (0.2 * x)**3 - np.cos(x) # функция

def derivative_func(x):
  return 0.024 * math.pow(x,2) + np.sin(x) # производная функции

def Newton(x, eps_2): # метод касательных
  y = x
  x = x - foo(x) / derivative_func(x)

  if (abs(y - x) >= eps_2):
    return Newton(x, eps_2)
  else:
    return x
```

```python
# Создаем массив значений x
x = np.linspace(-10, 10, 400)

# Вычисляем значения функции f(x)
y = foo(x)

# Строим график
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = (0.2 * x)^3 - cos(x)', color='blue')
plt.title('График функции f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.grid()
plt.legend()
plt.show()
```

```python
eps_1 = 0.001
eps_2 = 0.000001

print("МЕТОД ДИХОТОМИИ")

a = float(input("Введите значение a: "))
b = float(input("Введите значение b: "))

print(f"Значения a и b: {a} ; {b}")

while (abs(b - a) > eps_1):
  c = (a + b) / 2
  if (foo(b) * foo(c) < 0):
    a = c
  else:
    b = c
print(f"Корень уравнения равен: {c}")
```

```python
import numpy as np

# Начальные данные
n = 3
a = np.array([
    [14.38, -2.41, 1.39],
    [1.84, 25.36, -3.31],
    [2.46, -3.49, 16.37]
], dtype=float)

b = np.array([5.86, -2.28, 4.47], dtype=float)
x = np.zeros(n, dtype=float)
s = 0

# Печать начальной матрицы
print("\nНачальный вид матрицы:\n")
for i in range(n):
    for r in range(n):
        print(f"{a[i, r]:.2f}", end="\t")
    print()

print("\nСвободные члены матрицы:")
for i in range(n):
    print(f"B[{i + 1}] = {b[i]}")

# Прямой ход метода Гаусса
for k in range(n - 1):
    for i in range(k + 1, n):
        factor = a[i, k] / a[k, k]
        for j in range(k + 1, n):
            a[i, j] -= a[k, j] * factor
        b[i] -= b[k] * factor

# Обратный ход для нахождения корней
for k in range(n - 1, -1, -1):
    s = sum(a[k, j] * x[j] for j in range(k + 1, n))
    x[k] = (b[k] - s) / a[k, k]

# Вывод корней
print("\nМЕТОД ГАУССА, корни:")
for i in range(n):
    print(f"x{i + 1} = {x[i]:.5f}")

# Оценка невязки при разных точностях
for l in range(2, 5):
    x_rounded = [round(x[i], l) for i in range(n)]
    nevyazka = np.zeros(n, dtype=float)

    # Расчет невязки для каждого уравнения
    func = b[0] - (14.38 * x_rounded[0] - 2.41 * x_rounded[1] - 1.39 * x_rounded[2])
    func1 = b[0] - (14.38 * x[0] - 2.41 * x[1] - 1.39 * x[2])
    nevyazka[0] = func1 - func

    func = b[1] - (1.84 * x_rounded[0] - 25.36 * x_rounded[1] - 3.31 * x_rounded[2])
    func1 = b[1] - (1.84 * x[0] - 25.36 * x[1] - 3.31 * x[2])
    nevyazka[1] = func1 - func

    func = b[2] - (2.46 * x_rounded[0] - 3.49 * x_rounded[1] - 16.37 * x_rounded[2])
    func1 = b[2] - (2.46 * x[0] - 3.49 * x[1] - 16.37 * x[2])
    nevyazka[2] = func1 - func

    # Суммарная невязка
    func = sum(nevyazka)
    print(f"Невязка при {l} знаках = {abs(func):.5f}")
```

**Лабораторная работа 6. Вычисление интегралов**

```python
a = 0
b = 1
n = 10

# объявление функции
def f(x):
  return (3 * math.cos(x)) / (2 * x + 1.7)

# метод прямоугольников
def pryam(a, b, n):

  # объявление переменных
  sum = 0

  # вычисление шага
  dd = (b - a) / n

  # расчет суммы
  for i in range(n):
    sum += f(a + dd * i + dd / 2)

  # вычисление интеграла
  S = dd * sum
  return S

# метод трапеций
def trap(a, b, n):

  # объявление переменных
  sum = 0
  dd = (b - a) / n

  # расчет суммы
  for i in range(1, n):
    sum += f(a + dd * i)

  # вычисление интеграла
  S = dd * (f(a) / 2 + f(b) / 2 + sum)
  return S

# метод Симпсона
def simpson(a, b, n):

  # объявление переменных
  sum1 = 0
  sum2 = 0
  dd = (b - a) / n

  # расчет суммы по нечётным значениям x
  for i in range(1, n, 2):
    sum1 += f(a + dd * i)

  # расчет суммы по чётным значениям x
  for i in range(2, n - 1, 2):
    sum2 += f(a + dd * i)

  # вычисление интеграла
  S = (dd / 3) * (f(a) + 4 * sum1 + 2 * sum2 + f(b))
  return S

# метод Гаусса
def gauss(a, b, n):

  # объявление переменных
  sum = 0
  dd = (b - a) / n

  # расчет суммы
  for i in range(n):
    sum += f(a + dd * i + dd / 2 - dd / (2 * math.sqrt(3.0))) + f(a + dd * i + dd / 2 + dd / (2 * math.sqrt(3.0)))

  # вычисление интеграла
  S = dd / 2 * sum;
  return S
```

```python
print("ВЫЧИСЛЕНИЕ ИНТЕГРАЛА")
print("По формуле прямоугольников: ")
print(pryam(a, b, n))
print("По формуле трапеций: ")
print(trap(a, b, n))
print("По формуле Симпсона: ")
print(simpson(a, b, n))
print("По формуле Гаусса: ")
print(gauss(a, b, n))
```
